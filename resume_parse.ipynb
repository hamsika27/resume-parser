```python
import fitz  # PyMuPDF

# Get uploaded file name
filename = list(uploaded.keys())[0]

# Read PDF
doc = fitz.open(filename)
resume_text = ""
for page in doc:
    resume_text += page.get_text()
print(resume_text[:1000])  # preview first 1000 characters
```

```python
import spacy
import re

nlp = spacy.load("en_core_web_sm")

skills_db = [
    'python', 'java', 'machine learning', 'deep learning', 'excel',
    'data science', 'sql', 'c++', 'html', 'css', 'javascript',
    'django', 'flask', 'power bi', 'tableau', 'pandas', 'numpy'
]
```

```python
import re

# ✅ Clean and combine resume text into a single line (remove line breaks)
combined_text = ' '.join(resume_text.split())

# ✅ Extract Name from top few lines
lines = resume_text.split('\n')
cleaned_lines = [line.strip() for line in lines if line.strip()]
name = None
for line in cleaned_lines[:10]:
    if re.match("^[A-Za-z ]+$", line) and 2 <= len(line.split()) <= 4:
        name = line.strip()
        break
if not name:
    name = "Not Found"

# ✅ Extract Email
email_match = re.findall(r'\b\S+@\S+\.\S+\b', combined_text)
email = email_match[0] if email_match else "Not Found"

import re
# Combine all lines into one line
flat_text = ' '.join(resume_text.split())

# ✅ Phone Extraction (new, tested pattern)
phone_match = re.search(r'(\+?\d{1,3}[-\s.]?)?(\(?\d{3}\)?[-\s.]?\d{3}[-\s.]?\d{4})', flat_text)
phone = phone_match.group() if phone_match else "Not Found"

# ✅ Extract Skills
found_skills = []
for skill in skills_db:
    if skill.lower() in resume_text.lower():
        found_skills.append(skill)
```

```python
# ✅ Display results
print("✅ Parsed Resume Info:")
print(f"Name: {name}")
print(f"Email: {email}")
print(f"Phone: {phone}")
print(f"Skills: {', '.join(found_skills)}")
```
